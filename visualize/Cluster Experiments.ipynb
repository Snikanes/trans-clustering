```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

```python
def get_norm_stats(pts):
    total = 0.0
    total_count = (pts.shape[0] * (pts.shape[0] + 1)) / 2
    #print(total_count)
    min_norm = float('inf')
    max_norm = 0.0

    for i in range(0, pts.shape[0]):
        for j in range(i+1, pts.shape[0]):
            norm = np.linalg.norm(pts[i] - pts[j])
            if norm < min_norm:
                min_norm = norm
            elif norm > max_norm:
                max_norm = norm 
            total += norm
    return min_norm, max_norm, total / total_count 
```

```python
num_points = 1000
max_dim = 50
step = 1

for i in range(1, max_dim, step):
    rand_points = np.random.rand(num_points, i)
    min_norm, max_norm, mean_norm = get_norm_stats(rand_points)
    print("Dimensions {}".format(i))
    print("Min: {}".format(min_norm))
    print("Max: {}".format(max_norm))
    print("Ratio: {}".format(max_norm / min_norm))
    print("Mean: {}".format(mean_norm))
    print()
```
